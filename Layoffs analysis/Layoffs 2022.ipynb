```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib import style
style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
data = pd.read_csv('layoffs.csv')
data.head(10)
```

```python
data.info()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
data
```

```python
data.isnull().sum()
```

```python
avg_total_laid_off=data.total_laid_off.mean()
avg_percentage_laid_off=data.percentage_laid_off.mean()
avg_funds_raised=data.funds_raised.mean()
freq_industry=data.industry.mode()[0]
freq_stage=data.stage.mode()[0]

data.total_laid_off.replace(np.nan, avg_total_laid_off, inplace=True)
data.percentage_laid_off.replace(np.nan, avg_percentage_laid_off, inplace=True)
data.funds_raised.replace(np.nan, avg_funds_raised, inplace=True)
data.industry.replace(np.nan, freq_industry, inplace=True)
data.stage.replace(np.nan, freq_stage, inplace=True)

data
```
